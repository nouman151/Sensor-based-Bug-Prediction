```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn import metrics
```

**K Means clustring**

```python
k = 2 #dividing into 2 clusters(0 and 1)

# Initialize the KMeans model
kmeans = KMeans(n_clusters=k, random_state=40)

# Fit the model to the scaled data
kmeans.fit(df)

# Get the cluster labels
labels = kmeans.labels_

# Add the cluster labels to the original DataFrame
df['k-means'] = labels

# Print the DataFrame with cluster labels
print("\nDataFrame with Cluster Labels:")
print(df)
```

```python
# Reassign the 'bug' column back to the DataFrame
df['bug'] = bug_column
```

```python
df.head()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

actual = df["bug"]
predicted = df["k-means"]

confusion_matrix = metrics.confusion_matrix(actual, predicted)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [0, 1])

cm_display.plot()
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
accuracy = accuracy_score(actual, predicted)
precision = precision_score(actual, predicted)
recall = recall_score(actual, predicted)
f1 = f1_score(actual, predicted)
conf_matrix = confusion_matrix(actual, predicted)

# Print metrics
print(f"\nAccuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
print(f"Confusion Matrix:\n{conf_matrix}")
```

**DBSCAN**

```python
df=pd.read_csv("con_df.csv")
#dropping label ("bug") column from dataset and save it in another variable
bug_column = df.pop('bug')
df.shape
```

```python
from sklearn.cluster import DBSCAN
# DBSCAN parameters
eps = 0.3  # Maximum distance between two samples to be considered as part of the same neighborhood
min_samples = 5  # Minimum number of samples in a neighborhood for a data point to be considered a core point

# Initialize the DBSCAN model
dbscan = DBSCAN(eps=eps, min_samples=min_samples)

# Fit the model to the scaled data
dbscan.fit(df)

# Get the cluster labels (0 or 1)
dbscan_labels = dbscan.labels_

# Map cluster labels to {0, 1}
cluster_mapping = {label: 0 if label == dbscan_labels[0] else 1 for label in dbscan_labels}
dbscan_labels = [cluster_mapping[label] for label in dbscan_labels]

# Add the DBSCAN cluster labels to the DataFrame
df['dbscan_cluster'] = dbscan_labels
```

```python
# Reassign the 'bug' column back to the DataFrame
df['bug'] = bug_column
df.head()
```

```python
import matplotlib.pyplot as plt
import numpy
from sklearn import metrics

actual = df["bug"]
predicted = df["dbscan_cluster"]

confusion_matrix = metrics.confusion_matrix(actual, predicted)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [0, 1])

cm_display.plot()
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
accuracy = accuracy_score(actual, predicted)
precision = precision_score(actual, predicted)
recall = recall_score(actual, predicted)
f1 = f1_score(actual, predicted)
conf_matrix = confusion_matrix(actual, predicted)

# Print metrics
print(f"\nAccuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
print(f"Confusion Matrix:\n{conf_matrix}")
```

**Guassian Mixture Model- Model based Clustring**

```python
from sklearn.mixture import GaussianMixture
df=pd.read_csv("con_df.csv")
#dropping label ("bug") column from dataset and save it in another variable
bug_column = df.pop('bug')
df.shape
```

```python
# Number of clusters
k = 2

# Initialize the Gaussian Mixture Model
gmm = GaussianMixture(n_components=k, random_state=42)

# Fit the model to the scaled data
gmm.fit(df)

# Get the cluster labels
gmm_labels = gmm.predict(df)

# Add the GMM cluster labels to the DataFrame
df['gmm_cluster'] = gmm_labels

# Reassign the 'bug' column back to the DataFrame
df['bug'] = bug_column
df.head()
```

```python
import matplotlib.pyplot as plt
import numpy
from sklearn import metrics

actual = df["bug"]
predicted = df["gmm_cluster"]

confusion_matrix = metrics.confusion_matrix(actual, predicted)

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [0, 1])

cm_display.plot()
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
accuracy = accuracy_score(actual, predicted)
precision = precision_score(actual, predicted)
recall = recall_score(actual, predicted)
f1 = f1_score(actual, predicted)
conf_matrix = confusion_matrix(actual, predicted)

# Print metrics
print(f"\nAccuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
print(f"Confusion Matrix:\n{conf_matrix}")
```

DBSCAN and K-Modes gave best results among all of given clustring techniques.


1.   **DBSCAN** gave best results because we map the values to cluster 1, if it came into other than cluster 0.

2.   **K-Modes** itself gave best results(Accuracy and precision) because it calculate cost of each iteration and then improve learning parameter for next iteration, this is the reason that took more than 40 seconds to cluster the given df.
